```python
import cv2
import matplotlib.pyplot as LeHoangHiep
```

```python
img = cv2.imread('hinh1.png') # đọc ảnh
img
#OpenCV đọc ảnh dưới dạng ma trận NumPy với thứ tự màu BGR
```

```python
#Chiều của ảnh
print(img.shape)
```

```python
LeHoangHiep.imshow(img)
#Ảnh bị sai màu :xanh => đỏ
#do OpenCV dùng BGR, còn matplotlib dùng RGB

```

```python
# Cách 1: đảo chiều matrix
LeHoangHiep.imshow(img[:, :, ::-1])

```

```python
# Cách 2:convert các chế độ màu
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
LeHoangHiep.imshow(img_rgb)

```

```python
img_resize = cv2.resize(img_rgb, None, fx=1.5, fy=1.5)
#phóng to  nhỏ bằng hàm cv2.resize(), fx là chiều ngang, fy là chiều dọc
LeHoangHiep.imshow(img_resize)

```

```python
img_crop = img_rgb[100:300, 200:400]
LeHoangHiep.imshow(img_crop)

```

```python
img2 = cv2.imread('hinh2.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

```

```python
img2 = cv2.resize(img2, (img_rgb.shape[1], img_rgb.shape[0]))
```

```python
# ghép ngang
img_h = cv2.hconcat([img_rgb, img2])
LeHoangHiep.imshow(img_h)

```

```python
#ghép dọc
img_v = cv2.vconcat([img_rgb, img2])
LeHoangHiep.imshow(img_v)


```
